```python
import pandas as pd
import os

# ── Configuration ─────────────────────────────────────────────────────────────
INPUT_FILE = "MUP_DPR_RY25_P04_V10_DY23_NPIBN.csv"
OUTPUT_FILE = "pharmaguard_sample.csv"
SAMPLE_SIZE = 200000

# ── Loading ───────────────────────────────────────────────────────────────────
print("Loading dataset... (this may take 1-2 minutes)")
df = pd.read_csv(INPUT_FILE, low_memory=False)
print(f"Dataset loaded: {len(df):,} rows, {len(df.columns)} columns")
print(f"\nAvailable columns:\n{list(df.columns)}")

# ── Stratified sampling ───────────────────────────────────────────────────────
print(f"\nExtracting stratified sample of {SAMPLE_SIZE:,} rows...")

if 'Prscrbr_State_Abrvtn' in df.columns:
    sample = df.groupby('Prscrbr_State_Abrvtn', group_keys=False).apply(
        lambda x: x.sample(min(len(x), int(SAMPLE_SIZE * len(x) / len(df))))
    )
    sample = sample.sample(min(SAMPLE_SIZE, len(sample)), random_state=42)
else:
    sample = df.sample(n=min(SAMPLE_SIZE, len(df)), random_state=42)

print(f"Sample extracted: {len(sample):,} rows")

# ── Save ──────────────────────────────────────────────────────────────────────
sample.to_csv(OUTPUT_FILE, index=False)
size_mb = os.path.getsize(OUTPUT_FILE) / (1024 * 1024)
print(f"\nFile saved: {OUTPUT_FILE} ({size_mb:.1f} MB)")
print("Ready for the prototype!")
```
