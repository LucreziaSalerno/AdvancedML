```python
import pandas as pd

# Load sample
df = pd.read_csv("pharmaguard_sample.csv", low_memory=False)
print(f"Rows loaded: {len(df):,}")
```

```python
# Select relevant columns 
cols = [
    'Prscrbr_NPI',
    'Prscrbr_State_Abrvtn',
    'Prscrbr_Type',
    'Brnd_Name',
    'Gnrc_Name',
    'Tot_Clms',
    'Tot_Drug_Cst',
    'Tot_Benes',
    'Tot_Day_Suply'
]

df = df[cols].copy()
```

```python
# Rename columns
df.rename(columns={
    'Prscrbr_NPI':          'prescriber_id',
    'Prscrbr_State_Abrvtn': 'state',
    'Prscrbr_Type':         'specialty',
    'Brnd_Name':            'brand_name',
    'Gnrc_Name':            'generic_name',
    'Tot_Clms':             'total_claims',
    'Tot_Drug_Cst':         'total_cost',
    'Tot_Benes':            'total_patients',
    'Tot_Day_Suply':        'total_days_supply'
}, inplace=True)
```

```python
# Cleaning
# Remove nulls in key numeric columns
df.dropna(subset=['total_claims', 'total_cost', 'total_patients', 'total_days_supply'], inplace=True)

# Remove zero or negative values
df = df[(df['total_claims'] > 0) & (df['total_cost'] > 0) & (df['total_patients'] > 0)]

# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill remaining nulls in text columns
df['specialty'] = df['specialty'].fillna('Unknown')
df['brand_name'] = df['brand_name'].fillna(df['generic_name'])

```

```python
# Save clean dataset 
df.to_csv("pharmaguard_clean.csv", index=False)

print(f"Rows after cleaning: {len(df):,}")
print(f"Columns: {list(df.columns)}")
print(f"\nNull values remaining:\n{df.isnull().sum()}")
print(f"\nSample data:")
print(df.head(3))
```
